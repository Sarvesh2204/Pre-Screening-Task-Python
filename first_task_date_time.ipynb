```python
import inflect # package for converting numbers into 1st, 2nd, 3rd and so on
position = inflect.engine()
num = float(input("How much data you want to enter?: "))
dates = [] # stores the dates entered
times = [] # stores the times entered
i = 0
while i < num:
    date = input("Please enter a date? ")
    time = input("Please enter a time? ")
    dates.append(date)
    times.append(time)
    i+= 1
print("Thank you very much I will notify them!")

print("...")

for i in range(len(dates)):
    print(f"The {position.ordinal(i+1)} date has been reached! ({dates[i]} - {times[i]})")
```
